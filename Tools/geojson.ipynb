```python
import geopandas
from sqlalchemy import create_engine
```

```python
gdf = geopandas.read_file("../data/highway_michelstadt.geojson")
```

```python
sub = gdf[["osm_id", "highway", "geometry"]]
sub.head()
```

```python
engine = create_engine("postgresql://postgres:password@localhost:5432/osm")
sub.to_postgis("highway_michelstadt", engine) 
```

```python
sub.plot()
```
